```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.tsv',sep='\t')
```

```python
dev = pd.read_csv('data/dev.tsv',sep='\t')
```

```python
train = train[train['label'] == 1]
dev = dev[dev['label'] == 1]
```

```python
train_vals = train.values[:,1:-1]
eval_vals = train.values[:,1:-1]
```

```python
all_vocab = []
```

```python
for i in train_vals:
    for j in i:
        all_vocab.extend(j.split())
```

```python
for i in eval_vals:
    for j in i:
        all_vocab.extend(j.split())
```

```python
all_vocab = list(set(all_vocab))
```
